## Allocating the Data to Variables

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
print data.head()
X = data[data.columns.drop('Survived')]
result = data.Survived
```

## Cleaning the Data

```python
data.dtypes # we need to change all important categorical (objects) features to numerical format
data['Gender'] = data['Sex'].map( {'female': 0, 'male': 1} ).astype(int) # add gender to our dataframe

# Check which features have missing data
data.info() # we see that age, Cabin and embarked have less than 891 non-null entries

# fill in null ages with median age for each gender for each class
median_ages = np.zeros((2,3))

for i in range(2):
    for j in range(3):
        median_ages[i,j] = data[(data['Gender'] == i) & (data['Pclass'] == j+1)]['Age'].dropna().median()

data['AgeFill'] = data['Age']
for i in range(2):
    for j in range(3):
        data.loc[(data['Age'].isnull()) & (data['Gender'] == i) & (data['Pclass'] == j+1), 'AgeFill'] = median_ages[i,j]

# Repeat same procoess for Embarked feature (since there are only two, replace with 'S')
data.loc[(data['Embarked'].isnull()), 'Embarked'] = 'S'
data['EmbarkedNum'] = data['Embarked'].map( {'C': 1, 'Q': 2, 'S':3} ).astype(int)
```

## Exploring the Data

```python
import matplotlib.pyplot as plt
```

```python
def plotRelationship(X, y, var1):
    
    featureSurviveCount = {}
    for i in range(len(X)):
        feature = X[var1][i]
        featureSurviveCount[feature] = featureSurviveCount.get(feature, 0) + 1
    sortedFeatureSurviveCount = sorted(featureSurviveCount.iteritems(), key=lambda t: t[0], reverse=False)
    
    keys = []
    for key in range(len(sortedFeatureSurviveCount)):
        keys.append(sortedFeatureSurviveCount[key][0])
    values = []
    for value in range(len(sortedFeatureSurviveCount)):
        values.append(sortedFeatureSurviveCount[value][1])

    plt.figure(figsize=(50, 20))
    plt.title("Number of Survivors based on %s" % var1, fontsize = 60)
    plt.xlabel('%s'% var1, fontsize = 50)
    plt.ylabel('Number of Survivors', fontsize = 50)
    plt.rc('xtick', labelsize=30) 
    plt.rc('ytick', labelsize=50) 
    plt.bar(range(len(sortedFeatureSurviveCount)), values, align='center')
    plt.xticks(range(len(sortedFeatureSurviveCount)), keys)
    locs, labels = plt.xticks() #gets labels
    plt.setp(labels, rotation=60) #sets rotation of the labels
    plt.show()    
```

```python
features = ['Pclass', 'Age', 'Sex', 'SibSp', 'Parch', 'Fare', 'Embarked']
for feature in features:
    plotRelationship(X, result, feature)
```
