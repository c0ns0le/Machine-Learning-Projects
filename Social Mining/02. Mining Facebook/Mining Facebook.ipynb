### Examining Friendships

```python
friends = g.get_connections('Me', 'friends')['data']
```

```python
for friend in friends:
    print friend
```

```python
likes = {friend['name']: g.get_connections(friend['id'], 'likes')['data'] for friend in friends}
```

```python
friends_likes = Counter([like['name'] for friend in likes for like in likes[friend] if like.get('name')])
```

```python
pt = PrettyTable(field_names=['Name', 'Freq'])
pt.align['Name'], pt.align['Freq'] = 'l', 'r'
[ pt.add_row(fl) for fl in friends_likes.most_common(10) ]
print 'Top 10 likes amongst friends' 
print pt
```

```python
friends_likes_categories = Counter([like['category'] for friend in likes for like in likes[friend] if like.get('name')])
```

```python
pt = PrettyTable(field_names=['Name', 'Freq'])
pt.align['Name'], pt.align['Freq'] = 'l', 'r'
[ pt.add_row(fl) for fl in friends_likes_categories.most_common() ]
print 'Top categories for likes amongst friends' 
print pt
```

### Mutual Friendships

```python
import networkx as nx
import requests
```

```python
friends = {friend['id']: friend['name'] for friend in g.get_connections('me', 'friends')['data']}
```

```python
friends
```

```python
base_url = 'https://graph.facebook.com/me'
fields = 'friends{friends}'
url = '%s?fields=%s&access_token=%s' % (base_url, fields, ACCESS_TOKEN)
mutual_friends = getResults(url)
```

```python
mutual_friends_dict = {}

for item in mutual_friends['friends']['data']:
    friend_name = friends[item['id']]
    mutual_friends_dict[friend_name] = [mf['name'] for mf in item['friends']['data']]
```

```python
nxg = nx.Graph()
```

```python
[ nxg.add_edge(f1, f2) for f1 in mutual_friends_dict for f2 in mutual_friends_dict[f1] ]
print nxg.edges()
```

```python
cliques = [c for c in nx.find_cliques(nxg)]
print cliques
```

```python
num_cliques = len(cliques)
clique_sizes = [len(c) for c in cliques]
max_clique_size = max(clique_sizes)
avg_clique_size = sum(clique_sizes)/float(num_cliques)
max_cliques = [c for c in cliques if len(c) == max_clique_size]
king = list(reduce(lambda x, y: set(x).intersection(y), max_cliques))

print "Number of Cliques: ", num_cliques
print "Max Clique Size: ", max_clique_size
print "Average Clique Size: ", avg_clique_size
print "Biggest Cliques: ", max_cliques
print "King of Circles: ", king
```
