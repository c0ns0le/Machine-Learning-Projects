## Feature Engineering

```python
# Dates <-- edit test.csv Dates to be in the same format as train.csv Dates

# Time (hour) 
def getHour(Date):
    return int(Date[11:13])
df['Hour'] = df.Dates.map(lambda x: getHour(x))
```

```python
# Year
def getYear(Date):
    return int(Date[:4])
df['Year'] = df.Dates.map(lambda x: getYear(x))
```

```python
# Month
def getMonth(Date):
    return int(Date[5:7])
df['Month'] = df.Dates.map(lambda x: getMonth(x))

dummies_df = pd.get_dummies(df.Month) # and we can also create dummy variables from the quartiles
dummies_df = dummies_df.rename(columns=lambda x: 'Month_' + str(x))
df = pd.concat([df, dummies_df], axis=1)
```

```python
# Week in Year
def getWeek(Date):
    month = getMonth(Date)
    day = int(Date[8:10])
    week = int((month-1) * 4 + (np.floor(day/7)+1))
    return week
df['Week'] = df.Dates.map(lambda x: getWeek(x))
```

```python
# Day in Year (Holidays may have an effect)
def getDayInYear(Date):
    days = [0,31,59,90,120,151,181,212,243,273,304,334]
    month = getMonth(Date)
    day = int(Date[8:10])
    dayInYear = int(days[month-1] + day)
    return dayInYear
df['DayInYear'] = df.Dates.map(lambda x: getDayInYear(x))   
```

```python
# Day of the Week
df['DayOfWeekNum'] = pd.factorize(df.DayOfWeek)[0]

dummies_df = pd.get_dummies(df.DayOfWeek) # and we can also create dummy variables from the quartiles
dummies_df = dummies_df.rename(columns=lambda x: 'Day_' + str(x))
df = pd.concat([df, dummies_df], axis=1)
```

```python
# Weekend
def isWeekend(Day):
    if ((Day == 'Friday') or (Day == 'Saturday') or (Day == 'Sunday')):
        return 1
    else: return 0
df['Weekend'] = df.DayOfWeek.map(lambda x: isWeekend(x))
```

```python
# PdDistrict
df.PdDistrict = pd.factorize(df.PdDistrict)[0]
```

```python
# Address

# Corner
def isCorner(Address):
    if "/" in Address:
        return 1
    else: return 0
df['Corner'] = df.Address.map(lambda x: isCorner(x))
```

```python
# Street Name 
def getStreet(Address):
    words = Address.split(" ")
    return words[-2]
df['Street'] = df.Address.map(lambda x: getStreet(x))
df['StreetNum'] = pd.factorize(df.Street)[0]
```

```python
df['X_quart'] = pd.qcut(df.X, 4)
df['X_quart'] = pd.factorize(df.X_quart)[0]
df['Y_quart'] = pd.qcut(df.Y, 4)
df['Y_quart'] = pd.factorize(df.Y_quart)[0]
```

```python
# New Years Eve, Superbowl Sunday, Black Friday
def isNewYears(Date):
    month = getMonth(Date)
    day = int(Date[8:10])
    if ((month == 12) and (day == 31)):
        return 1
    else: return 0
    
def isSuperBowl(Date):
    month = getMonth(Date)
    day = int(Date[8:10])
    if ((month == 2) and  (day < 8)): # first Sunday of february
        return 1
    else: return 0
    
def isBlackFriday(Date):
    month = getMonth(Date)
    day = int(Date[8:10])
    if ((month == 11) and (day > 23)): # Last friday of Nov unless Thanksgiving is 23rd
        return 1
    else: return 0

df['NewYears'] = df.Dates.map(lambda x: isNewYears(x))
df['isSuperBowl'] = df.Dates.map(lambda x: isSuperBowl(x))
df['isBlackFriday'] = df.Dates.map(lambda x: isBlackFriday(x))
```

```python
def isDay(day, Day):
    if (day == Day):
        return 1
    else: return 0
def isMonth(monthNum, Date):
    month = getMonth(Date)
    if (monthNum == month):
        return 1
    else: return 0
    
df['isSunday'] = df.DayOfWeek.map(lambda x: isDay("Sunday", x))
df['isFeb'] = df.Dates.map(lambda x: isMonth(2, x))
df['SuperBowl'] = df.isSuperBowl * df.isSunday * df.isFeb

df['isFriday'] = df.DayOfWeek.map(lambda x: isDay("Friday", x))
df['isNov'] = df.Dates.map(lambda x: isMonth(11, x))
df['BlackFriday'] = df.isBlackFriday * df.isFriday * df.isNov
```

```python
# Drop the object variable and useless extracted features except category (results)
# print df.info()

df = df.drop(['Dates', 'Descript', 'DayOfWeek', 'PdDistrict', 'Resolution', 'Address', 'Street', 'X', 'Y', 
              'isSuperBowl', 'isBlackFriday', 'isSunday', 'isFeb', 'isFriday', 'isNov'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df.Hour = scaler.fit_transform(df.Hour)
df.Year = scaler.fit_transform(df.Year)
df.Month = scaler.fit_transform(df.Month)
df.Week = scaler.fit_transform(df.Week)
df.DayInYear = scaler.fit_transform(df.DayInYear)
df.DayOfWeekNum = scaler.fit_transform(df.DayOfWeekNum)
df.StreetNum = scaler.fit_transform(df.StreetNum)
df.X_quart = scaler.fit_transform(df.X_quart)
df.Y_quart = scaler.fit_transform(df.Y_quart)
```
